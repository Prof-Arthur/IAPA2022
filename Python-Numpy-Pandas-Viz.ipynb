```python
import numpy as np
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

# lineplot (data)
# barplot (x, y)
# heatmap (data, annot)
# kdeplot
# histplot
# scatter
# regplot (regression line)
```

```python
# Lendo dados de um arquivo .csv
dados = pd.read_csv('sample_data/california_housing_train.csv')
# print(dados)

# Usando pyplot para definir as dimensões da figura
plt.figure(figsize=(14,6))

# Mudando o estilo do gráfico com o Seaborn
# sns.set_style('darkgrid')

# Line Plot com o Seaborn
sns.lineplot(data=dados['population'][:80])

# Definindo Título, Label do eixo X, Label do eixo Y
plt.title("Título")
plt.xlabel('X')
plt.ylabel('Y')
```

```python
# Criando uma matrix 10x10 com valores aleatórios seguindo uma distribuição normal com média 50 e variância 5
dados = np.random.randn(10,10) * 5 + 50
# Criando um Data Frame com a matrix gerada
df = pd.DataFrame(dados)

# Usando pyplot para definir o tamanho da figura
plt.figure(figsize=(10,7))
# Usando Seaborn para plotar um Heatmap com os dados gerados
sns.heatmap(data=dados, annot=True)
```

```python
# Definindo a quantidade de dados amostrados por cada distribuição
# Vamos começar com valores pequenos e ir crescendo para verificarmos o efeito do tamanho na amostragem aleatória
n_data = 10

# Amostrando aleatoriamente a partir de uma distribuição uniforme (0 -> 1)
unif = np.random.rand(n_data)
# Amostrando aleatoriamente a partir de uma distribuição normal (média: 0, variância: 1)
norm = np.random.randn(n_data)

# Definindo dimensões da figura
plt.figure(figsize=(20,7))

# Subdividindo figura em duas partes (e plotando na parte 1 - lado esquerdo)
plt.subplot(1, 2, 1)
# Histograma da distribuição uniforme
sns.histplot(unif)
plt.title('Distribuição Uniforme')

# Subdividindo figura em duas partes (e plotando na parte 2 - lado direito)
plt.subplot(1,2,2)
# Histograma da distribuição normal
sns.histplot(norm)
plt.title('Distribuição Normal')
```

```python
# Criando dados sintéticos com a seguinte relação: y = f(x) = a*x + b + ruido_aleatório
a = 2
b = 5
ruido = 3
f = lambda x: a * x + b + ruido * np.random.randn()

# Amostrando X com valores também aleatórios (de uma distribuição uniforme)
n_dados = 50 # Quantidade de pontos
amp_x = 20 # Amplitude de variação dos dados (em vez de variar de 0 a 1)
X = amp_x * np.random.rand(n_dados)
# Gerando Y a partir de X utilizando a função f
Y = list(map(f, X))

# Tamanho da figura
plt.figure(figsize=(12, 7))
# Scatter plot usando o Seaborn
sns.scatterplot(x = X, y = Y)
# Reg plot usando o Seabborn
# sns.regplot(x = X, y = Y)
```
